#Task 1 — Explore and Identify Issues

```python
import pandas as pd

df = pd.read_csv("/content/used_cars_messy.csv")

print(df.shape)
print(df.info())
print(df.describe(include='all'))
```

#Task 2 — Clean the Data

```python
df = df.dropna(subset=['selling_price'])
df
```

```python
df['brand'] = df['brand'].str.strip().str.lower()
df['brand']
```

```python
df['mileage'] = df['mileage'].str.extract('(\d+\.?\d*)').astype(float)
df['mileage']
```

```python
df['mileage'].fillna(df['mileage'].median(), inplace=True)
df.fillna(df.mode().iloc[0], inplace=True)
```

```python
df = df.drop_duplicates()
df
```

#Task 3 — Compute Baseline MAE

```python
from sklearn.metrics import mean_absolute_error

mean_price = df['selling_price'].mean()

df['baseline_pred'] = mean_price

mae = mean_absolute_error(df['selling_price'], df['baseline_pred'])

print("Baseline MAE:", mae)
```
